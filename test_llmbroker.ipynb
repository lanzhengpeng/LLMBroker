```python
import openai
client = openai.OpenAI(
    base_url="http://127.0.0.1:8000/v1",  # 你的本地服务地址
    api_key="",  # 如果没有鉴权，就传空字符串
)

response = client.chat.completions.create(
     model="GLM-4-Flash",
    messages=[{
        "role": "user",
        "content": "你可以做什么"
    }
              ],
    stream=False
)

print("模型回复：", response.choices[0].message.content)
```
